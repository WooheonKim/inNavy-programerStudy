- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
>  
```
>>> print(c) > 10

```

````python
class Counter() :
    def __init__(self, count, step = 1) :
        self.count = count
        self.step = step
    def __str__(self) :
        s = "[count (step : %d)] %d" % (self.step, self.count)
        return s
    def incr(self) :
        self.count += self.step
    def __call__(self) :
        self.count += self.step
    def __add__ (self, n) :
        self.count += n
        return self
    def __sub__ (self, n) :
        self.count -= n
        return self
    def __lt__(self, n) :
        return self.count < n
    def __le__(self, n) :
        return self.count <= n
    def __eq__(self, n) :
        return self.count == n
    def __ne__(self, n) :
        return self.count != n
    def __gt__(self, n) :
        return self.count > n
    def __ge__(self, n) :
        return self.count >= n
    
a = Counter(10)
b = Counter(10, 5)
print(a)
print(b)
a.incr()
b.incr()
print(a)
print(b)
a()
b()
print(a)
print(b)
a = a - 5
b = b - 5
print(a)
print(b)
print(a < b)
print(a <= b)
print(a == b)
print(a != b)
print(a > b)
print(a >= b)
````

[4번 문제]

다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.  
- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

````python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
````

__init__ 은 클래스의 생성자 메소드이다. 클래스 객체를 선언할 때 리스트 인자 l를 같이 넣어줄 경우 l 안의 원소를 클래스 객체 내의 인자로 포함시킨다. 그 후 MySet.eliminate_duplicate(self) (뒤에서설명) 메소드를 통해 클래스 내부의 중복되는 인자를 삭제한다.

__str__ 은 클래스의 출력형태를 만들어주는 메소드이다. "MySet: {"이라는 문자열에 클래스 내 객체를 차례대로 추가해 반환하는 작업을 하고있다.

eliminate_duplicate 메소드는 객체 안의 인자를 하나씩 다른 리스트에 삽입하여 만일 삽입 도중에 이미 리스트 내에 해당 원소가 있다면 그 원소는 추가로 삽입하지 않는다. 그 후 원래 리스트를 비우고 중복이 제거된 리스트 s의 내용물을 원래 리스트 l에 삽입한다.

- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.  
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
>  
```
>>> u = s | t

```

- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
>  
```
>>> u = s & t 
>>> print(u)

```
- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
>  
```
>>> s = MySet([1, 2, 3]) 
>>> t = MySet([3, 4, 5]) 
>>> u = s - t 
>>> print(u) 
MySet: {1, 2} 
```

````python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    def __or__(self, other) :
        list = self + other # 두 집합을 더한 후
        MySet.eliminate_duplicate(list) # 중복인자를 제거한다.
        list.sort() #  정렬한다.
        return list
        
    def __and__(self, other) :
        list = [] # 새 리스트를 만든 후
        for e in other:
            if e in self :
                list.append(e) # other내부의 요소 e가 self 내부에 있다면 새 리스트에 해당 요소를 추가한다.
        MySet.eliminate_duplicate(list) # 중복인자를 제거한다.
        return list
    def __sub__(self, other) :
        list = [] # 새 리스트를 만든 후
        for e in other :
            if e not in self : # self의 원소중 other의 원소와 겹치지 않는 원소가 있다면
                list.append(e) # list에 추가한다.
        return list
        
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    
    a = s | t
    print(a)
    a = s & t
    print(a)
    a = s - t
    print(a)
````
