 1). 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s1.txt에 있는 파일 내용은 다음과 같다.
 
    cat dog
    dog pig
    ham bird
    pig ham

```python
with open('s.txt', 'r') as f:
    print("Before :")
    list = []
    for line in f:
        print(line, end='') #라인단위로 출력
        left = line.split()[0] #라인의 첫번째단어
        right = line.split()[1] #라인의 두번째단어
        list.append((left, right)) #리스트에 포함
    list.sort()
    sortFile = ""
    for l in list:
        sortFile += "%s %s\n" % (l[0], l[1])
with open('s1.txt', 'w') as f: f.write(sortFile)
with open('s1.txt', 'r') as f:
    print("\nAfter :")
    print(f.read())
```

 2). 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.

    ham bird
    cat dog
    pig ham
    dog pig

```python
with open('s.txt', 'r') as f:
    print("Before :")
    list = []
    for line in f:
        print(line, end='')
        left = line.split()[0]
        right = line.split()[1]
        list.append((right, left)) #라인단위로 출력한 내용을 두번째 단어를 기준으로 정렬하기 위해 두번째부터 삽입.
    list.sort()
    sortFile = ""
    for l in list:
        sortFile += "%s %s\n" % (l[1], l[0])
with open('s2.txt', 'w') as f: f.write(sortFile)
with open('s2.txt', 'r') as f:
    print("\nAfter :")
    print(f.read())
```

3).파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s3.txt에 있는 파일 내용은 다음과 같다.

    pig ham cat
    dog ham bird
    dog pig

```python
with open('s.txt', 'r') as f:
    print("Before :")
    list = []
    for line in f:
        print(line, end='')
        for i in line.split(): list.append(i) # 리스트에 순서대로 삽입.
    sortFile = ""
    for i in range(len(list)):
        if i % 3 == 0 and i != 0 : #3개씩 끊어 구분.
            sortFile += "\n"
        sortFile += "%s " % list[i]
with open('s3.txt', 'w') as f: f.write(sortFile)
with open('s3.txt', 'r') as f:
    print("\nAfter :")
    print(f.read())
```

```python
import hashlib

def Menu(): #메뉴를 부르고 값을 입력받아 추가조작을 하는 함수
    return int(input("""
Welcome to Our Service
  1. Sign Up
  2. Sign In
  3. Quit
"""))

def userInfo(s, Type): #유저정보를 확인, 비교하기 위해 사용하는 함수
    if Type == 'id': return s.split(':')[0]
    elif Type == 'password' : return s.split(':')[1].split(',')[0].strip()
    elif Type == 'name' : return s.split(':')[1].split(',')[1].strip()
    elif Type == 'school' : return s.split(':')[1].split(',')[2].strip()
def encryptPassword(s):
    return hashlib.sha1(s.encode()).hexdigest()

menu = Menu()
while(1):
    if menu == 3: break
    elif menu == 1:
        dup = False
        id = input("아이디를 입력하세요 : ")
        with open("access.txt", "r") as f:
            for line in f:
                if id == userInfo(line, 'id'): #아이디 중복확인
                    print("Sorry, the entered ID is already used.")
                    dup = True 
                    break
        if dup == True : continue # 아이디가 중복이면 처음부터 다시
        password = input("비밀번호를 입력하세요 : ")
        name = input("이름을 입력하세요 : ")
        school = input("학교명을 입력하세요 : ")
        encrypted_Password = encryptPassword(password) #비밀번호 암호화
        with open("access.txt", "a") as f: #파일을 추가모드로 오픈
            f.write("%s: %s, %s, %s\n" % (id, encrypted_Password, name, school)) #파일에 추가
        print('\n' + '=' * 30)
    elif menu == 2:
        id = input("아이디를 입력하세요 : ")
        password = input("비밀번호를 입력하세요 : ")
        encrypted_Password = encryptPassword(password)
        existID= False #아이디가 파일 내에 존재하는 지 체크하는 변수
        with open("access.txt", "r") as f: #파일을 read 모드로 오픈
            for line in f: #한 줄씩 읽으며
                if id == userInfo(line, 'id') and encrypted_Password == userInfo(line, 'password'): #아이디와 비밀번호 모두 파일 내에 있다면
                    print("Hello %s!" % userInfo(line, 'name'))
                    existID = True
                    break
                elif id == userInfo(line, id) and encrypted_Password != userInfo(line, password): #아이디가 존재하고, 비밀번호가 틀렸을 때
                    while encrypted_Password != userInfo(line, password): #비밀번호 재입력
                        print("Sorry, the entered password is not correct.")
                        encrypted_Password = encryptPassword(input("비밀번호를 입력하세요 : "))
                    print("Hello %s!" % userInfo(line, 'name'))
                    existID = True
                    break
            if existID == 0 : print("Sorry, you are not a registered member.") #아이디가 파일 내에 없을 때
            print('\n' + '=' * 30)
    menu = Menu()
```

3. range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
> * range(stop)
> * range(start, stop[, step])

2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.

3) frange 사용 예
* 3-1) frange(0.5)
    * [0.0, 0.1, 0.2, 0.3, 0.4]
* 3-2) frange(1.0, 2.0)
    * [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
* 3-3) frange(2.2, 4.0, 0.5)
    * [2.2, 2.7, 3.2, 3.7]

```python
def frange(start = 0.0, stop = None, step = 0.1) :
    if stop == None :
        stop = start
        start = 0.0
    rangeList = []
    result = []
    while start < stop:
        rangeList.append(start)
        start = start + step
    stepDigit = 1
    while step * 10 % 1 != 0.0 :
        stepDigit = stepDigit + 1
        step = step * 10
    if (stepDigit == 1) :
        return ["%.1f" % i for i in rangeList]
    if (stepDigit == 2) :
        return ["%.2f" %  i for i in rangeList]
    if (stepDigit == 3) :
        return ["%.3f" %  i for i in rangeList]
print(frange(0.5))
print(frange(1.0, 2.0))
print(frange(2.2, 4.0, 0.5))
print(frange(1.0, 2.0, 0.05))
```

sum() 사용 예
* sum()
    * 0
* sum(1, 2)
    * 3
* sum(1, 2, 3, 4, 5)
    * 15
* sum(1, 5, 7, 2, -10)
    * 5

```python
def sum(*args): # 가변인수를 받는 함수 sum
    result = 0
    for i in args: result += i # 가변인수의 크기만큼 반복
    return result
print(sum())
print(sum(1, 2))
print(sum(1, 2, 3, 4, 5))
print(sum(1, 5, 7, 2, -10))
```

-------------------------------
* 1). 다음에 제시되는 함수들을 모두 이용해야 한다.
    * split
    * map
    * join

* 2) myinitial() 함수 사용 예
    * myinitial("as soon as possible")
    * ASAP

```python
def firstUpper(s):
    return s[0].upper()
def myinitial(s):
    return "".join(map(firstUpper, s.split())) # 문자열이 split으로 분리, firstUpper를 통해 분리된 각 단어의 첫글자가 대문자로 변경, join으로 합체
print(myinitial("as soon as possible"))
```

6. 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

[옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def myfact(n):
    if n == 1: return 1
    return n * myfact(n-1) 
print(myfact(2965))
```

```python
def saveHTMLFile(url):
    with open("%s.html" % url.split('//')[1], "w", encoding='utf-8') as f:
        f.write(requests.get(url).text) #url 내부의 내용을 파일에 저장
```

```python
def filterTextAndSaveFrequencyData(url):
    req = requests.get(url)
    source = re.sub(r'\n', ' ', html.unescape(req.text)) # 줄바꿈 제거
    source = re.sub(r'<script>', '', source) # script 태그 제거
    source = re.sub(r'</script>', '', source) 
    source = re.sub(r'<style>', '', source) # style 태그 제거
    source = re.sub(r'</style>', '', source)
    source = re.sub(r'<!(--)?.*?-->', '', source) #html 주석 제거
    source = re.sub(r'</?.*?>', '\n', source) #HTML 태그 제거
    line = source.split('\n')
    countWords = {}
    for i in line:
        if i.strip() == '': continue; #공백밖에 없으면 넘어감
        words = [r for r in i.strip().split()] #단어 사이의 whitespace를 기준으로 파싱
        for r in range(len(words)) :
            for p in string.punctuation: # string 모듈의 구두문자 리스트에서
                words[r] = words[r].replace(p, '') # 구두문자를 비교해 찾아 제거
            if words[r].strip() == ' ' : continue # 공백은 세지 않음.
            if words[r] in countWords : countWords[words[r]] += 1 #단어가 있다면 갯수추가
            else: countWords[words[r]] = 1 # 없다면 단어 추가
    for k in countWords.keys():
        if k in Abbreviation : countWords[k] = -1 #불용어는 카운트제외
    with open("%s.words_frequency.pickle" % url.split('//')[1], "wb") as f: pickle.dump(countWords, f) #피클 파일 저장
```
